#### The data show how visitors to (or time spent in) categorized places change compared to baseline days. A baseline day represents a normal value for that day of the week. The baseline day is the median value from the 5-week period Jan 3 – Feb 6, 2020.
#### To make the reports useful, categories have been used to group some of the places with similar characteristics for purposes of social distancing guidance. The following categories are available:
##### - retail and recreation, i.e. places like restaurants,cafes, shopping centers, theme parks,museums, libraries, and movie theaters
##### - grocery and pharmacy, i.e. grocery markets, food warehouses, farmers markets, specialty food shops, drug stores, and pharmacies
##### - parks, i.e. national parks, public beaches, marinas, dog parks, plazas,and public gardens
##### - transit stations i.e. all public transport hubs such as subway, bus, and train stations
##### - workplaces, i.e. places of work
##### - residential, i.e. people’s residence

# EXERCISE 2- Random number generators

#### • one of the first random number generator was proposed by von Neumann, the so-called middle square algorithm

#### • write R code to implement this type of generator and, given a fixed digit number input, square it an remove the leading and trailing digits, in order to return a number with the same number of digits as the original number

#### • Suggestion : after having squared the number, convert it to a list of characters  (number <- unlist(strsplit(as.character(x.squared),"")))  and, after having removed the head and tail of the list, convert it back to a number (as.numeric(paste(number.after.trimming, collapse="")))


## Case 1: Random generation from a number that change each time

In this case we are going to consider the method for a number that change each time i.e.  each generated number is generated from a random number that is not linked to the prevision one. In this way it is possible to fix the digits and to get a generated number with the same digits of the starting one

```r
MSM<- function(x) {
    #list with starting number and his square
    list_x<-unlist(strsplit(as.character(x),""))
    list_xsq<-unlist(strsplit(as.character(x^2),""))
    
    #lenght to compute the condition to get mid digits
    x_length<-length(list_x)
    xsq_length<-length(list_xsq)
    difference <- xsq_length- x_length
        
    
     if (difference%%2==0){
         x <- (paste(list_xsq[(difference/2+1):(xsq_length-(difference/2))], collapse=""))
     }
    else {
        x <- (paste(list_xsq[((difference+1)/2):(xsq_length-((difference+1)/2))], collapse=""))
    }
         return(x)
  
}
x<-sample(100000:999999,20)

print(paste(' starting number',x," ", '  (starting number)^2:',x^2, " " ,  '  generated number:',sapply(x,MSM))) 
```

## Case 2 : generation from a fixed number 

In this case we are going to start from a fixed number and we are going to generate the numbers as explained in "AdvStat 4 PhysAna - Stat 04b" by professor A. Garfagnini. 

In particular the Algorithm is:
- take a number with a large number of digits and square it 
- extract the central digits
- repeat the sequence from 1

```r
MSM<- function(x,N) {

    cat( "the starting number is", x , "\n")
    cat( "the sequence generated is:" , "\n")
    for(i in 0:N){
        
    list_x<-unlist(strsplit(as.character(x),""))
    xsq<-x^2
    list_xsq<-unlist(strsplit(as.character(xsq),""))
    x_length<-length(list_x)
    xsq_length<-length(list_xsq)
    difference <- xsq_length- x_length
    z<-x
  if (difference%%2==0){
      x <- (paste(list_xsq[(difference/2+1):(xsq_length-(difference/2))], collapse=""))
      }
    else
    {
        x <- (paste(list_xsq[((difference+1)/2):(xsq_length-((difference+1)/2))], collapse="")
             )}
   y<- x
        print(paste0(i,"th generation:", " " , ' seed:', " ", z," ", '  (seed)^2:'," " ,  xsq, " " ,  '  generated number:', " " ,y)) 
         x<- as.numeric(paste(x, collapse=""))
        
  }
  
}
MSM(5772156649,20)
```

In this case is possible to see that the digits of the generated number decrease after $n$=18 generations from a fixed number. In fact it is possible to have a 0 as first digit so the number must lose the first digit ( in this case 058725929 $\longrightarrow$ 587259294) and this is why we lose digits from computation.

## Case 3: generation from a fixed number and failure of the algorithm

In this case we see how the algorithm fails from same sequences. In particular this shows how the middle square algorith is not the best one to generate random numbers (in fact we can have multiple 0 or 1 in a row so the square could be 0 or 1) and it invites us to use other methods more accurate.

```r
MSM<- function(x,N) {

    cat( "the starting number is", x , "\n")
    cat( "the sequence generated is:" , "\n")
    for(i in 0:N){
        
    list_x<-unlist(strsplit(as.character(x),""))
    xsq<-x^2
    list_xsq<-unlist(strsplit(as.character(xsq),""))
    x_length<-length(list_x)
    xsq_length<-length(list_xsq)
    difference <- xsq_length- x_length
    z<-x
  if (difference%%2==0){
      x <- (paste(list_xsq[(difference/2+1):(xsq_length-(difference/2))], collapse=""))
  }
  else {
        x <- (paste(list_xsq[((difference+1)/2):(xsq_length-((difference+1)/2))], collapse=""))
    }
    y<- x
    print(paste0(i,"th", "  generation:", " " , ' starting number:', " ", z," ", '  (starting number)^2:'," " ,  xsq, " " ,  '  generated number:', " " ,y)) 
    x<- as.numeric(paste(x, collapse=""))
  }
  
}
MSM(6829,20)
```

# EXERCISE 3- Bayesian Inference

 ##  A publishing company has recently launched a new journal. In order to determine how effective it is in reaching its possible audience, a market survey company selects a random sample of people from a possible target audience and interviews them. Out of 150 interviewed people, 29 have read the last issue of the journal.


## 1) What kind of distribution would you assume for y, the number of people that have seen the last issue of the yournal ?

#### I would assume a $ \small{Binomial Distribution} $ for $ y $. In fact if I pick randomly one person from the set the probability that she has read the journal is $p$ while the probability she hasn't $ q= 1- p $. Hence, taking a set of $n$ people,  the  number of them that has read the journal follows a binomial distribution with parameters $n$ and $p$.

## 2) Assuming a uniform prior, what is the posterior distribution for y ?



#### - Let's adopt a uniform prior, $ P(p|M) \sim U(0,1) $

#### - the Posterior pdf is simply proportional to the Likelihood
#### $$
P(p|y,n,M)= \frac{1}{Z} p^{y} (1-p)^{n-y}
$$

#### - where $Z$ is the normalitation factor (i.e. the evidence $ P(y|n,M) $)

## 3) Plot both posterior and likelihood ditributions functions

```r
n<- 150
y<- 29
n.sample <-2000
delta.p<- 1/n.sample
p<- seq( from=1/(2*n.sample), by=1/n.sample , length.out=n.sample)
post.unif <- dbinom( x=29, size=150,prob=p)
likelihood <- dbinom( x=29, size=150,prob=p)
p.mean<- delta.p*sum(p*p.norm)

#plot
options(repr.plot.width=13, repr.plot.height=10)
par(mfrow=c(2,2))
plot(p, post.unif/(delta.p*sum(post.unif)),
    col="deeppink",
    type="l",
    lwd=3,
    las=1,
    xlab="p",
    ylab= " P(p|y,n,M) ",
     main= "Posterior PDF "
    )

p.norm.u<- post.unif/(delta.p*sum(post.unif))
p.mean.u<- delta.p*sum(p*p.norm.u)
abline(v=p.mean.u,col="black",lty=4,lwd=2)

#legend
legend(x = "topright",
       col = c("black"),
       lty = 4,
       lwd = 2,
       cex=0.75,
       legend = c('Mean'))



#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="lightgray",
     lwd=1) 


plot(p, likelihood,
    col="seagreen1",
    type="l",
    lwd=3,
    las=1,
    xlab="p",
    ylab= " P(y|p,n,M)",
     main= "Likelihood PDF "
    )


#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="lightgray",
     lwd=1) 


#plot likelihood vs posterior
options(repr.plot.width=15, repr.plot.height=12)

par(mar= c(5,4,4,4)+0.3)

plot(p,p.norm.u,
     type="l",
     col="deeppink", 
     las=1,
     lwd=3,
     lty=4,
     xlab="p",
     ylab="",
     main="  Posterior PDF vs Likelihood PDF  with 2 y-axis")

par(new=TRUE)

plot(p,likelihood,
     type="l", 
     col="seagreen1", 
     las=1,
     lwd=3,
     lty=3,
     axes=FALSE,
     xlab=" ",
     ylab="P(p|y,n,M)",)



axis(side=4,at=pretty(range(likelihood)),las=1)
mtext("P(y|p,n,M)",side=4,line=2.5)

#legend
legend(x = "topright",
       col = c("seagreen1", "deeppink"),
       lty = 3,
       lwd = 4,
       legend = c('Likelihood', 'Posterior')) 

#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="gray",
     lwd=1) 

plot(p,p.norm.u,
     type="l",
     col="deeppink", 
     las=1,
     lwd=3,
     lty=1,
     xlab="p",
     ylab="Density",
     main="  Posterior PDF vs Likelihood PDF ")
lines(p,likelihood,
     type="l", 
     col="seagreen1", 
     las=1,
     lwd=3,
     lty=1
    )


#legend
legend(x = "topright",
       col = c("seagreen1", "deeppink"),
       lty = 1,
       lwd = 4,
       legend = c('Likelihood', 'Posterior')) 

#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="gray",
     lwd=1) 
```

## 1) Assuminga flat prior,and a beta prior,plot the likelihood,prior and posterior distributions for the data set.


```r
#prior
alpha <- 15
beta <- 15
n<-30
n.sample<-2000
delta.p <- 1/n.sample
p <- seq(from=1/(2*n.sample), by=1/n.sample, length.out=n.sample)
r<-15
p.prior <- dbeta(p, alpha, beta)
unif.prior<-dunif(p, min = 0, max = 1, log = FALSE)

options(repr.plot.width=13, repr.plot.height=10)
par(mfrow=c(2,3))

plot( p,p.prior,
     col="navy",
     type="l",
     lty=1,
     lwd = 3,
     ylim=c(0,5),
     xlab="p",
     ylab="P(p|M)",
     main="Priors")

lines(p, unif.prior,
      type="l",
      lwd=3,
      lty=1,
      col="darkred",
      xlim=c(0,1),
      ylim=c(0,1.1))

#legend
legend(x = "topright",
       col = c("navy", "darkred"),
       lty = 1,
       lwd = 2,
       cex=0.75,
       legend = c('Beta', "Flat"))
#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="lightgray",
     lwd=1) 



#likelihood
p.star <- dbinom(x=r, size=n, prob=p)
plot(p,p.star,
     type="l",
     lwd=1,
     lty=2,
     main="Likelihood",
     col="black",
    ylab="Density",
     xlab="p")

#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="lightgray",
     lwd=1) 


#Posteriors
p.norm <- p.star/(delta.p*sum(p.star))
post.beta<-dbeta(x=p,alpha+r,beta+n-r)

plot(p, post.beta,
     col="navy",
     type="l",
     lty=1,
     lwd = 3,
     ylim=c(0,7),
     xlab="p",
     ylab="P(p|n,r,M)",
     main="Posteriors")

lines(p, p.norm,
      type="l",
      lwd=3,
      lty=1,
      col="darkred")

#legend
legend(x = "topright",
       col = c("navy", "darkred"),
       lty = 1,
       lwd = 2,
       cex=0.75,
       legend = c('Beta', "Uniform"))
#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="lightgray",
     lwd=1) 

```

## 2) Evaluate the most probable value for the coin probability $ p $ and, integrating the posterior probability distribution, give an estimate for a 95% credibility interval.


As reported in "AdvStat 4 PhysAna - Stat 05" of professor A. Garfagnini if we start with a Beta Prior with parameters $ \alpha_ p $ and $ \beta_ p$, and then measure $ r $ heads in n tosses, the Posterior is a Beta functions with parameters:

$ \alpha =\alpha_p + r $ and $ \beta =\beta_p + n-r $ .

Furthermore a Uniform prior is a Beta prior with $ \alpha = 1 $ and $ \beta = 1 $

So we can use the quantile function to find the credibility interval.

```r
mp.value.post.unif <- which.max(p.norm)*(1/length(p))
mp.value.post.beta <- which.max(post.beta)*(1/length(p))
cat("The most probable value assuming an uniform prior is:" , mp.value.post.unif, "\n")
cat("The most probable value assuming a beta prior is:" , mp.value.post.beta, "\n")

cat("==============================================================================","\n")

#quantile for uniform-post
qunif5<-qbeta(0.025, 1+r, 1+n-r, ncp = 0, lower.tail = TRUE, log.p = FALSE)
qunif95<-qbeta(0.975, 1+r, 1+n-r, ncp = 0, lower.tail = TRUE, log.p = FALSE)

#quantile for beta-post
qbeta5<-qbeta(0.025, alpha+r, beta+n-r)
qbeta95<-qbeta(0.975, alpha+r, beta+n-r)

cat("The 95% credibility interval assuming an uniform prior is:" , "[", round(qunif5,3) , ", " ,round(qunif95,3),"]", "\n")
cat("The 95% credibility interval assuming a beta prior is:" , "[", round(qbeta5,3) , ", " ,round(qbeta95,3),"]", "\n")

```

```r

#plot
x <- seq(qunif5, qunif95, 0.01)
y <-  dbeta(x, 1+r, 1+n-r)
x2 = c(qunif5,x,qunif95)
y2 = c(0,y,0)

options(repr.plot.width=10, repr.plot.height=8)
par(mfrow=c(2,3))
plot( p,unif.prior,
     col="darkred",
     type="l",
     lty=1,
     lwd = 3,
     ylim=c(0,5),
     xlab="p",
     ylab="P(p|M)",
     main="Uniform-Prior")

#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="lightgray",
     lwd=1) 


plot(p, p.star,
     type="l",
     lwd=1,
     lty=2,
     main="Likelihood",
     col="darkred",
    ylab="Density",
     xlab="p")

#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="lightgray",
     lwd=1) 

plot(p, p.norm,
     col="darkred",
     type="l",
     lty=1,
     lwd = 3,
     ylim=c(0,6),
     xlab="p",
     ylab="P(p|n,r,M)",
     main="Posterior")
polygon(x2,y2, col="coral", border=NA)
abline(v=mp.value.post.unif, lty=2, col="black")
abline(v=qunif5, lty='dotted', col="red")
abline(v=qunif95, lty='dotted', col="red")

#legend
legend(x = "topright",
       col = c("darkred", "black"),
       lty = 1,
       lwd = 2,
       cex=0.5,
       legend = c('Posterior', "Mode"))
#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="lightgray",
     lwd=1) 

x <- seq(qbeta5, qbeta95, 0.01)
y <-  dbeta(x, alpha+r, alpha+n-r)
x2 = c(qbeta5,x,qbeta95)
y2 = c(0,y,0)

#plot
plot( p,p.prior,
     col="navy",
     type="l",
     lty=1,
     lwd = 3,
     ylim=c(0,5),
     xlab="p",
     ylab="P(p|M)",
     main="Beta-Prior")

#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="lightgray",
     lwd=1) 


plot(p, p.star,
     type="l",
     lwd=1,
     lty=2,
     main="Likelihood",
     col="navy",
    ylab="Density",
     xlab="p")

#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="lightgray",
     lwd=1) 

plot(p, post.beta,
     col="navy",
     type="l",
     lty=1,
     lwd = 3,
     ylim=c(0,7),
     xlab="p",
     ylab="P(p|n,r,M)",
     main="Posterior")
polygon(x2,y2, col="lightblue", border=NA)
abline(v=mp.value.post.unif, lty=2, col="black")
abline(v=qbeta5, lty='dotted', col="red")
abline(v=qbeta95, lty='dotted', col="red")

#legend
legend(x = "topright",
       col = c("navy", "black"),
       lty = 1,
       lwd = 2,
       cex=0.5,
       legend = c('Posterior', "Mode"))
#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="lightgray",
     lwd=1) 
```

## 3) Repeat the same analysis assuming a sequential analysis of the data. Show how the most probable value and the credibility interval change as a function of the number of coin tosses (i.e. from 1 to 30).


```r
#vector with results
tosses<-c("T","T","T","T","T","H","T","T","H","H","T","T","H","H","H","T","H","T","H","T","H","H","T","H","T","H","T","H","H","H")

#initial values
n<-1
n.sample<-2000
delta.p <- 1/n.sample
p <- seq(from=1/(2*n.sample), by=1/n.sample, length.out=n.sample)
alpha<-15
beta<-15
nHt<-0
nTt<-0
mp.value.post.unif <- c(NULL)
mp.value.post.beta <- c(NULL)
qunif5 <- c(NULL)
qbeta5 <- c(NULL)
qunif95 <- c(NULL)
qbeta95 <- c(NULL)

for(toss in tosses)
    {
    if(toss=="H") {
    nHt<-nHt+1
    }
   if(n<=30){
        
       #posteriors
         postunif <- dbeta(p, 1+nHt, 1+n - nHt)
         postbeta <- dbeta(p, alpha+nHt, beta+n-nHt)
      
       #most probable values
         mp.value.post.unif <-c(mp.value.post.unif, which.max(postunif)*(1/length(p)))
         mp.value.post.beta <-c( mp.value.post.beta, which.max(postbeta)*(1/length(p)))
         
         #quantile for credibility intervals
         qunif5<-c(qunif5,qbeta(0.025, 1+nHt, 1+n-nHt, ncp = 0, lower.tail = TRUE, log.p = FALSE))
         qunif95<-c(qunif95,qbeta(0.975, 1+nHt, 1+n-nHt, ncp = 0, lower.tail = TRUE, log.p = FALSE))

   
         qbeta5<-c(qbeta5,qbeta(0.025, alpha+nHt, alpha+n-nHt, ncp = 0, lower.tail = TRUE, log.p = FALSE))
         qbeta95<-c(qbeta95,qbeta(0.975, alpha+nHt, alpha+n-nHt, ncp = 0, lower.tail = TRUE, log.p = FALSE))
         
          }
     n<-n+1
     }

     
col <- rgb(0, 75, 0, max = 150, alpha = 20, names = "seagreen1")
col2 <- rgb(0,0 , 100, max = 100, alpha = 20, names = "blue")

 
#plot for most probable values and credibility interval
options(repr.plot.width=12, repr.plot.height=8)

x =1:30

plot(1:length(qunif95),
     qunif95,
     type='l',
     ylim=c(0,1),
     col='seagreen1',
     xlab='step',
     ylab='most probable p', 
     lty='dotted',
     lwd=2,
     main='Sequential analysis')

polygon(x=c(x,rev(x)),
        y=c(qunif95,rev(qunif5)),
        col = col,
        border = NA)
polygon(x=c(x,rev(x)),
        y=c(qbeta95, rev(qbeta5)),
        col = col2,
        border = NA)

lines(1:length(mp.value.post.unif),
      mp.value.post.unif,
      type='l',
      col='seagreen1',
      lwd=5)

lines(1:length(qunif5), 
      qunif5,
      type='l',
      col='seagreen1',
      lty='dotted',
      lwd=2)

lines(1:length(qbeta95),
      qbeta95,
      type='l',
      ylim=c(0,1),
      col='cyan1',
      lty='dotted',
      lwd=2)

lines(1:length(mp.value.post.beta),
      mp.value.post.beta,
      type='l', 
      col='cyan1', 
      lwd=5)

lines(1:length(qbeta5),
      qbeta5,
      type='l',
      col='cyan1',
      lty='dotted', 
      lwd=2)

#legend
legend(2, 1,
       legend=c("Most probable value (uniform prior)",
                "Most probable value (beta prior)",
                expression(paste("CL"["95%"], " for uniform prior")),
                expression(paste("CL"["95%"], " for beta prior"))),
       col=c("seagreen1",
             "cyan1",
             col,
             col2),
       lty=1:1,lwd=5:5)
#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="lightgray",
     lwd=1) 

#plot for sequential-analysis for posteriors
plot(p, postbeta,
     col="navy",
     type="l",
     lty=1,
     lwd = 3,
     ylim=c(0,7),
     xlab="p",
     ylab="Probability density",
     main="Final Posteriors- Sequential analysis")

lines(p, postunif,
      type="l",
      lwd=3,
      lty=1,
      col="darkred")

#legend
legend(x = "topright",
       col = c("navy", "darkred"),
       lty = 1,
       lwd = 2,
       cex=0.75,
       legend = c('Beta', "Uniform"))
#grid
grid(nx=NULL,
     ny=NULL,
     lty=3,
     col="lightgray",
     lwd=1) 

```
