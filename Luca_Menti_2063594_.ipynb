EXERCISE 3

- it is a binomial distribution with y=17 and n=145
- auming beta(a,b) prior we got beta posterior with $a1=a+y$ and $b1=b+n-y$

```r
#question 1:
cat("-It is a binomial distribution with y=17 and n=145","\n")

#question 2
y<-17
n<-145
p<-seq(0,0.2,0.001)
a<-1
b<-4

a1<-a+y
b1<-b+n-y

mean<-a1/(a1+b1)

var<-a1*b1/(((a1+b1)^2)*(a1+b1))
q1<-qbeta(0.025,a1,b1)
q2<-qbeta(0.975,a1,b1)

cat("the mean is:", mean,"\n")
cat("the var is:", var,"\n")

cat("-The 95% CI is: [", q1 , q2 ,"]")


post<-dbeta(p,a1,b1)
plot(p,post,type="l",col="red",main="Post assuming beta prior")
abline(v=c(q1,q2),col="blue")
legend(x="topleft",col=c("red","blue"),lwd=2,legend=c("post","CI"))
#normal
postn<-dnorm(p,mean=mean,sqrt(var))
plot(p,postn,type="l",col="red",main="Post assuming norm approx")

n1<-qnorm(0.025,mean,sqrt(var))
n2<-qnorm(0.975,mean,sqrt(var))

cat("-The 95% CI norm aprox is: [", n1 , n2 ,"]")


#normal
postn<-dnorm(p,mean=mean,sqrt(var))
plot(p,postn,type="l",col="red",main="Post assuming norm approx")
abline(v=c(n1,n2),col="blue")
legend(x="topleft",col=c("red","blue"),lwd=2,legend=c("post","CI"))
```

EXERCISE 6

- The posterior distribution assuming uniform prior is a gamma function with $\alpha$= $\sum$ y + 1 and $\lambda$=n

- the mean is m= $\frac{\alpha}{\lambda}$ and the variance var=$\frac{\alpha}{\lambda^2}$

```r
x<-c(5,8,4,6,11,6,6,5,6,4)
alphau=sum(x)+1
lambdau=length(x)
mu<-seq(3,12,0.001)
post<-dgamma(mu,alphau,lambdau)

mean=alphau/lambdau
var=alphau/lambdau^2
std=sqrt(var)
cat("the mean is:", mean,"\n")
cat("the std is:", std,"\n")
cat("the mode is:", qgamma(0.5,alphau,lambdau),"\n")

l1<-qgamma(0.025,alphau,lambdau)
l2<-qgamma(0.975,alphau,lambdau)

cat("-The 95% CI  is: [", l1 , l2 ,"]")

plot(mu,post,type="l",col="red")
abline(v=c(l1,l2),col="blue")
abline(v=mean,col="green")
legend(x="topleft",col=c("red","blue","green"),lwd=2,legend=c("post","CI","mean"))

```
